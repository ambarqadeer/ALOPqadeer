```python
import sp
from astropy.io import fits
import glob
import numpy as np
```

```python
sciencelist=sorted(glob.glob('Logs/Corrected_Science/r*.fit'))
sharpnesslist=np.zeros(len(sciencelist),dtype='float32')
for sciencec in sciencelist[0:11]:
    hdu=fits.open(sciencelist[sciencec])
    science=hdu[0].data
    find=sp.find(science,hmin=1000,fwhm=12)
    try:
        sharpnesslist[sciencec]=max(find[3])
        print(sciencen,'detected')        
    except:
        print(sciencen,'did not return any objects')
        continue
max(sharpnesslist)
```
